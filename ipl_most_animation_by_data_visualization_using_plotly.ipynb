```python
#the data visualization of fastest centuries of all seasons using plotly expression
# tips: take x=category(least),animation_frame=time based,y=continous,hover_name=category,range_y=first test and fix the range
```

```python
import plotly.express as px
```

```python
px.bar(df,x="Player",y="Runs",animation_frame="Match Date",hover_name="Against")
```

```python
px.bar(df,x="Player",y="Runs",animation_frame="Match Date",hover_name="Venue",range_y=(0,180))
```
